```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

```python
import yfinance as yf
import datetime
from datetime import date, timedelta

end_date = '2023-09-01'#today.strftime("%Y-%m-%d")
start_date = '2019-09-01'


bitcoin_data = yf.download('BTC-USD', 
                           start=start_date, 
                           end=end_date, 
                           progress=False)

ethereum_data = yf.download('ETH-USD', 
                           start=start_date, 
                           end=end_date, 
                           progress=False)
```

```python
data = bitcoin_data.values.reshape(-1, 1)

# Normalize the data
scaler = MinMaxScaler()
data = scaler.fit_transform(data)
```

```python
data
```

```python
def create_sequences(data, sequence_length):
    X, y = [], []
    for i in range(len(data) - sequence_length):
        X.append(data[i:i+sequence_length])
        y.append(data[i+sequence_length])
    return np.array(X), np.array(y)

sequence_length = 10 # Adjust as needed
X, y = create_sequences(data, sequence_length)
```

```python
train_size = int(0.8 * len(X))
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
model = keras.Sequential([
    keras.layers.LSTM(units=50, activation='relu', input_shape=(sequence_length, 1)),
    keras.layers.Dense(units=1)
])

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
```

```python
y_pred = model.predict(X_test)

# Inverse transform the predictions to get them back to the original scale
y_pred = scaler.inverse_transform(y_pred)
y_test = scaler.inverse_transform(y_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(y_test, label='Actual')
plt.plot(y_pred, label='Predicted')
plt.legend()
plt.title('Bitcoin Price Prediction')
plt.show()
```
